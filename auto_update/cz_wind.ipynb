```python
import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter

# 目前的这些文件路径都是之前实习生用绝对路径写的，以后有时间了一定要改成相对路径
data_folder_path = 'C:\\Users\\goodluck\\Desktop\\DB'
raw_data_path =  "C:\\Users\\goodluck\\Desktop\\DB\\raw_data"
storage_path =  "C:\\Users\\goodluck\\Desktop\\DB\\database_storage"
daily_path =  "C:\\Users\\goodluck\\Desktop\\DB\\daily"

# 这个是DTBS.pkl的文件地址
DTBS_path = os.path.join(storage_path, "DTBS.pkl")


# re-open，读取DTBS.pkl存放到DTBS中
with open(DTBS_path, 'rb') as f:  
    DTBS = pickle.load(f)
    
from WindPy import w
w.start()
```

```python
cds = []
died = []
ld = DTBS['D']['day'][-1]
print(ld)
for cd in DTBS['A'].keys():
    if DTBS['A'][cd][ld]['ia'] == 1:
        cds.append(cd)
    else:
        died.append(cd)
```

```python
lw = DTBS['D']['day'][-5]
print(lw)
start_time = lw
end_time = ld
index = 'strbvalue'
stock_code = cds.copy()
```

```python
out = w.wsd(stock_code, index, start_time, end_time)
```

```python
out_df = pd.DataFrame(out.Data,index=out.Codes,columns=out.Times)
out_df = out_df.T
out_df
```
