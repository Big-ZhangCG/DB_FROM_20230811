```python
# 这个文件主要是为了更新DTBS的B区中  hot_concept  字段
# 可转债所属的热门概念
# 数据来源： i问财网页爬虫 mywork_crawler_iwencai.ipynb 爬取到的数据存储到csv中，爬虫有可能会失效，i问财有反爬机制
# 处理方式： 读取上述csv并存储

import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter

data_folder_path = '../'
raw_data_path = "../raw_data"
storage_path = "../database_storage"
daily_path = "../daily"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")
DCBS_path = os.path.join(storage_path, "DCBS.pkl")
with open(DTBS_path, 'rb') as f:
    DTBS = pickle.load(f)
```

```python
bond_code_list = list(DTBS['B'].keys())

df = pd.read_csv('问财_所属概念.csv')
mapping = df.set_index('股票代码')['所属概念'].to_dict()
print(df)

for bond_code in bond_code_list:
    stock_code = DTBS['B'][bond_code]['sc']
    DTBS['B'][bond_code]['hot_concept'] = mapping.get(stock_code, 'i问财中无对应概念，可能原因是：对应正股已退市')
    
    # i问财中如果没有对应的热点概念，大概率是由于正股已经退市
    if DTBS['B'][bond_code]['hot_concept'] == 'i问财中无对应概念，可能原因是：对应正股已退市':
        print(bond_code, stock_code)
```
