```python
# 爬取问财上的数据
import pywencai
```

```python
# z这里是问题列表，需要添加的话在这里添加
question_list = ['可转债概念','所属概念']

# 在这里选择需要查询那些问题，直接输入问题的编号，全部查询的话输入0即可
print(f'可选择查询的问题（输入0代表全部查询，输入1代表查询{question_list[0]}，输入2代表查询{question_list[1]}，以此类推）：\n{question_list}')

query_mode = int(input('请输入正确的查询问题编号（可以输入0、1、2...）'))
```

```python
def query_and_save(query_question):
    # 调用pywencai库去爬取问题答案，并且将获取到的DataFrame保存到csv中
    
    print(query_question)
    
    # 核心功能
    res = pywencai.get(question=query_question, loop=True)
    print(res)
    print(f'{query_question}，爬取成功！')

    # 输出文件名称
    output_file_name = '问财_' + query_question + '.csv'
    res.to_csv(output_file_name, index=False, encoding='utf-8-sig')
    print(f'{query_question}，保存成功！')


if query_mode == 0:
    # 查询全部问题
    for query_question in question_list:
        # 逐个查询问题并依次保存到EXCEL中
        query_and_save(query_question)
else:
    # 查询单个问题并保存到EXCEL中
    query_question = question_list[query_mode-1]
    query_and_save(query_question)
```
