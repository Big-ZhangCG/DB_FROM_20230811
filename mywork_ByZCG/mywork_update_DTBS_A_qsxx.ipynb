```python
# 这个文件主要是为了修改DTBS的A区中  下修/强赎    存在的错误，过一段时间可以重新跑一次，这样有错误的话可以改过来
# 如果要得到最新的结果需要先运行mywork_update_DTBS_B_xr_xt_qr_qt文件 去获取所有新的转债的xr\xt1\xt2\qr\qt1\qt2信息

import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter

data_folder_path = '../'
raw_data_path = "../raw_data"
storage_path = "../database_storage"
daily_path = "../daily"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")
DCBS_path = os.path.join(storage_path, "DCBS.pkl")
with open(DTBS_path, 'rb') as f:  
    DTBS = pickle.load(f)
```

```python
date = DTBS['D']['day']

# 四舍五入的示例代码
from decimal import Decimal, ROUND_HALF_UP

def round_decimal(value, decimal_places):
    # 实现四舍五入功能
    rounded_value = Decimal(value).quantize(Decimal('0.1') ** decimal_places, rounding=ROUND_HALF_UP)
    return rounded_value


for code in DTBS['A'].keys():

    # 获取强赎触发比例，若不存在则返回130
    qr = DTBS['B'][code].get('qr', 130)
    print(qr)

    # 获取强赎生效区间，若不存在则返回15
    qr1 = DTBS['B'][code].get('qr1', 15)

    # 获取强赎生效最大区间，若不存在则返回30
    qr2 = DTBS['B'][code].get('qr2', 30)

    # 获取下修触发比例，若不存在则返回85
    xr = DTBS['B'][code].get('xr', 85)
    print(xr)
    # 获取下修生效区间，若不存在则返回15
    xr1 = DTBS['B'][code].get('xr1', 15)

    # 获取下修生效最大区间，若不存在则返回30
    xr2 = DTBS['B'][code].get('xr2', 30)
    
    s_code = DTBS['B'][code]['sc']
    
    this_code_dates_list = []
    
    for td in DTBS['A'][code].keys():
        
        # 判定存活
        if DTBS['A'][code][td]['ia'] == 1: 
            
            this_code_dates_list.append(td)
            
            # 获取当日qs
            # 为了准确，四舍五入后保留两位有效数字
            if DTBS['E'][s_code][td]['cl'] >= round_decimal(qr * 0.01 * DTBS['A'][code][td]['csp'], 2):
                DTBS['A'][code][td]['qs'] = 1
            else:
                DTBS['A'][code][td]['qs'] = 0

            # 获取当日xx
            if DTBS['E'][s_code][td]['cl'] < round_decimal(xr * 0.01 * DTBS['A'][code][td]['csp'], 2):
                DTBS['A'][code][td]['xx'] = 1
            else:
                DTBS['A'][code][td]['xx'] = 0
    
    sorted_this_code_dates_list = sorted(this_code_dates_list)
    
    for i in range(len(sorted_this_code_dates_list)):
        
        # 对于某一只转债的某一天情况而言，给xx15、xx30、qs15、qs30赋初值0
        xx15 = 0
        xx30 = 0
        qs15 = 0
        qs30 = 0
        
        if i < qr1 - 1:
            for j in range(0, i+1):
                xx15 = xx15 + DTBS['A'][code][sorted_this_code_dates_list[j]]['xx']
        else:
            for j in range(i-qr1+1, i+1):
                xx15 = xx15 + DTBS['A'][code][sorted_this_code_dates_list[j]]['xx']
                
        if i < qr2 - 1:
            for j in range(0, i+1):
                xx30 = xx30 + DTBS['A'][code][sorted_this_code_dates_list[j]]['xx']
        else:
            for j in range(i-qr2+1, i+1):
                xx30 = xx30 + DTBS['A'][code][sorted_this_code_dates_list[j]]['xx']
        
        
        if i < xr1 - 1:
            for j in range(0, i+1):
                qs15 = qs15 + DTBS['A'][code][sorted_this_code_dates_list[j]]['qs']
        else:
            for j in range(i-xr1+1, i+1):
                qs15 = qs15 + DTBS['A'][code][sorted_this_code_dates_list[j]]['qs']
                
        if i < xr2 - 1:
            for j in range(0, i+1):
                qs30 = qs30 + DTBS['A'][code][sorted_this_code_dates_list[j]]['qs']
        else:
            for j in range(i-xr2+1, i+1):
                qs30 = qs30 + DTBS['A'][code][sorted_this_code_dates_list[j]]['qs']
                
        
        DTBS['A'][code][sorted_this_code_dates_list[i]]['qs15'] = qs15
        DTBS['A'][code][sorted_this_code_dates_list[i]]['qs30'] = qs30
        DTBS['A'][code][sorted_this_code_dates_list[i]]['xx15'] = xx15
        DTBS['A'][code][sorted_this_code_dates_list[i]]['xx30'] = xx30
    
```

```python
# 保存上面的这些更改

f_save = open(DTBS_path, 'wb')
pickle.dump(DTBS, f_save)
f_save.close()
```
