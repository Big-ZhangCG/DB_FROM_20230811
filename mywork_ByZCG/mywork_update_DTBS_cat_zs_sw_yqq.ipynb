```python
# 这个文件主要是为了修改、更新DTBS的B区中  cat1,cat2,sw,zs,yqq   等行业分类指标，过一段时间可以重新跑一次。

import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter

data_folder_path = '../'
raw_data_path = "../raw_data"
storage_path = "../database_storage"
daily_path = "../daily"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")
DCBS_path = os.path.join(storage_path, "DCBS.pkl")

with open(DTBS_path, 'rb') as f:  
    DTBS = pickle.load(f)
```

```python
print(DTBS['B'])

bond_code_list = list(DTBS['B'].keys())
stock_code_list = [DTBS['B'][bond_code]['sc'] for bond_code in bond_code_list]
stock_code_set = list(set(stock_code_list))

print(len(bond_code_list),len(stock_code_list), len(stock_code_set))
```

```python
from WindPy import w
import datetime

# 获取今天的日期
today = datetime.date.today()
print('获取数据的日期：', today)

w.start()

# 用正股代码查询申万一级/二级分类
cat1_list = w.wsd(stock_code_set, "industry_sw_2021", today, today, "industryType=1").Data
cat2_list = w.wsd(stock_code_set, "industry_sw_2021", today, today, "industryType=2").Data

print(cat1_list,cat2_list)
```

```python
cat1_mapping = {stock_code_set[stock_code_index] : cat1_list[0][stock_code_index] for stock_code_index in range(len(stock_code_set))}
cat2_mapping = {stock_code_set[stock_code_index] : cat2_list[0][stock_code_index] for stock_code_index in range(len(stock_code_set))}
print(len(cat1_mapping.keys()),cat1_mapping)
```

```python
for bond_code in DTBS['B'].keys():
    # 获得对应正股及转债代码，添加cat1和cat2
    stock_code = DTBS['B'][bond_code]['sc']
    
    DTBS['B'][bond_code]['cat1'] = cat1_mapping[stock_code]
    DTBS['B'][bond_code]['cat2'] = cat2_mapping[stock_code]
```
