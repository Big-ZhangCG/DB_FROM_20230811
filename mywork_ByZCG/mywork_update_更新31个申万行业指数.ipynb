```python
# 这个文件主要是为了 修改/更新 DTBS的C区中 31个申万行业行业指数收盘价

import pandas as pd
import numpy as np
import pickle
import time 
import os
from datetime import datetime
from operator import itemgetter

data_folder_path = '../'
raw_data_path = "../raw_data"
storage_path = "../database_storage"
daily_path = "../daily"
DTBS_path = os.path.join(storage_path, "DTBS.pkl")
DCBS_path = os.path.join(storage_path, "DCBS.pkl")
with open(DTBS_path, 'rb') as f:  
    DTBS = pickle.load(f)
```

```python
# 找到需要添加的时间段，中证转债的时间列表 - 行业指数的时间列表 的结果就是我们需要补充的时间段
need_update_date_list = sorted(list(set(DTBS['C']['zi'].keys())-set(DTBS['C']['nlmy'].keys())))
if len(need_update_date_list) == 0:
    print('申万行业行业指数已全部更新，无需运行后面的代码块。')
else:
    print('申万行业行业指数需要更新，继续运行后面的代码块。')
    start_day = need_update_date_list[0]
    end_day = need_update_date_list[-1]
    print(start_day,end_day)
    print(DTBS['C'])
```

```python
from WindPy import w
w.start()

from datetime import datetime

# 我们需要获取的31个申万行业行业指数代码
sw_industry_index_codes = ['801010.SI','801030.SI','801040.SI','801050.SI','801080.SI','801110.SI','801120.SI','801130.SI','801140.SI','801150.SI',
                           '801160.SI','801170.SI','801180.SI','801200.SI','801210.SI','801230.SI','801710.SI','801720.SI','801730.SI','801740.SI',
                           '801750.SI','801760.SI','801770.SI','801780.SI','801790.SI','801880.SI','801890.SI','801950.SI','801960.SI','801970.SI','801980.SI']

# 对应的缩写
sw_industry_index_abbr = ['nlmy','jchg','gt','ysjs','dz','jydq','spyl','fzfs','qgzz','yysw',
                          'gysy','jtys','fdc','smls','shfw','zh','jzcl','jzzs','dlsb','gfjg',
                          'jsj','cm','tx','yx','fyjr','qc','jxsb','ms','sysh','hb','mrhl']

# 对应的名称
sw_industry_index_names = ['农林牧渔','基础化工','钢铁','有色金属','电子','家用电器','食品饮料','纺织服饰','轻工制造','医药生物',
                           '公用事业','交通运输','房地产','商贸零售','社会服务','综合','建筑材料','建筑装饰','电力设备','国防军工',
                           '计算机','传媒','通信','银行','非银金融','汽车','机械设备','煤炭','石油石化','环保','美容护理']


# 调用wind函数去获取行业指数收盘价
informations = w.wsd(sw_industry_index_codes, "close", start_day, end_day, "").Data

print(informations)
```

```python
print(len(informations),len(informations[0]))

# 在C区对应位置赋值
for i in range(len(sw_industry_index_abbr)):
    
    index_abbr = sw_industry_index_abbr[i]
    
    for j in range(len(need_update_date_list)):
        
        need_update_date = need_update_date_list[j]
        
        DTBS['C'][index_abbr][need_update_date] = informations[i][j]
        
print(DTBS['C'])
```
